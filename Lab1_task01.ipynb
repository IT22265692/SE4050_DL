```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

plt.hist(exp_data, bins=100, density=True, histtype='step', label='Exponential')
plt.hist(uni_data, bins=100, density=True, histtype='step', label='Uniform')
plt.hist(norm_data, bins=100, density=True, histtype='step', label='Normal')

plt.axis([-2.5, 10, 0, 1.2])
plt.legend()
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Plot of Z = X² + Y²")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Assuming df is your DataFrame
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_sub = df[features]

# Pearson
pearson_corr = df_sub.corr(method='pearson')
# Spearman
spearman_corr = df_sub.corr(method='spearman')

# Heatmap for Pearson
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")
plt.show()

# Heatmap for Spearman
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation")
plt.show()
```
